```python
import pandas as pd
import matplotlib.pyplot as plt

#Criando o conjunto de dados
dados = {
    'Cidades': ['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Salvador', 'Manaus'],
    'Temperaturas': [28.5, 34.2, 22.0, 31.8, 33.5],
    'Umidade': [60, 75, 55, 80, 90]
}

df = pd.DataFrame(dados)
print("Dados carregados com sucesso!")
df
```

```python
#Calculando métricas importantes
temp_media = df['Temperaturas'].mean()
cidade_quente = df.loc[df['Temperaturas'].idxmax(), 'Cidades']
cidade_fria = df.loc[df['Temperaturas'].idxmin(), 'Cidades']

print(f"Temperatura média geral: {temp_media:.2f}°C")
print(f"Cidade mais quente: {cidade_quente}")
print(f"Cidade mais fria: {cidade_fria}")
```

```python
#Criando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(df['Cidades'], df['Temperaturas'], color=['skyblue','orange','lightgreen','red','darkred'])

#Customização
plt.title('Comparativos de temperaturas por Cidades', fontsize=14)
plt.xlabel('Cidades')
plt.ylabel('Temperaturas (ºC)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

#Exibindo
plt.show()
```

#Análise dos Resultados:

Observamos uma amplitude térmica significativa entre Curitiba (22ºC) e Rio de Janeiro (34.2°) evidenciando a diferença climática regional.

Cidades como Manaus, e Salvador apresentam alta temperatura aliada a alta umidade (acima de 80%), o que sugere uma sensação térmica muito superior à registrada.

Recomendação: Campanhas de hidratação e atenção à saúde devem ser priorizadas no Rio de Janeiro e Manaus baseando-se nestes dados.

```python
import pandas as pd
import numpy as np

#Simulando dados reais com problemas
dados_sujos = {
    'Cidades': ['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Salvador', 'Manaus'],
    'Temperaturas': [28.5, 34.2, 22.0, np.nan, 335], # Salvador sem dado, Manaus com erro (335ºC)
    'Umidade': [60, 75, 55, 80, 90]
}

df_limpeza = pd.DataFrame(dados_sujos)
print("Dados com 'sujeira':")
df_limpeza
```

```python
#Verifica quantos valores nulos existem por coluna
print(df_limpeza.isnull().sum())
```

```python
#Usando a Média para não perder dados. Outra solução: Excluir a linha para não considerar
media_temp = df_limpeza['Temperaturas'].mean()
df_limpeza['Temperaturas'] = df_limpeza['Temperaturas'].fillna(media_temp)
```

```python
# Corrigindo manualmente o valor absurdo
df_limpeza.loc[df_limpeza['Temperaturas'] > 100, 'Temperaturas'] = 33.5
```
